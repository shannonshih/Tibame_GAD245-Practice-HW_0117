```python
data.groupby('Platform')[['NA_Sales','Global_Sales']]. \
aggregate([np.sum,np.max,np.min])
```

```python
data.groupby('Platform')[['NA_Sales','Global_Sales']].aggregate(
    [lambda x: x.max() - x.min(),
    lambda x: x.median() - x.mean()]
)
```

```python
def A1(x):
    return np.max(x) - np.min(x)
def B2(x):
    return np.median(x) - np.mean(x)
data.groupby('Platform')[['NA_Sales','Global_Sales']].aggregate(
    [A1,B2]
)
```

## Transform

```python
data.groupby('Platform')['Global_Sales'].transform(sum)
```

```python
data.groupby('Platform')['Global_Sales'].aggregate(np.sum)
```

## apply

```python
data2 = data.copy()
global_sum = data2.groupby('Platform')['Global_Sales'].transform(np.sum)
data2['銷售佔比'] = data2['Global_Sales'] / global_sum
data2
```

```python
#使用apply來達成等同上面的功能
def rate(x):
    x['銷售佔比'] = x['Global_Sales'] / x['Global_Sales'].sum()
    return x
data2 = data.copy()
data2.groupby('Platform').apply(rate)
```

## filter

```python
# 列出該平台總銷售額大於 1000 百萬的遊戲
index = data.groupby('Platform')['Global_Sales'].transform(sum) > 1000
data[index]
```

```python
# 列出所有PS平台的遊戲
def AAA(x):
    slist = ['PS','PS2','PS3']
    return any(x.iloc[0]['Platform'] in s for s in slist)
data.groupby('Platform').filter(AAA)
```
